```python
columnas = ['views', 'likes', 'dislikes', 'comment_count']

# Crea un diccionario para almacenar el recuento de outliers por cada columna
proporcion_outliers = {}

for col in columnas:
    Q1 = dataset[col].quantile(0.25)
    Q3 = dataset[col].quantile(0.75)
    IQR = Q3 - Q1

    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    outliers = dataset[(dataset[col] < limite_inferior) | (dataset[col] > limite_superior)][col]

    proporcion= len(outliers) / len(dataset)
    proporcion_outliers[col] = proporcion

    # No reemplaces los valores, simplemente cu√©ntalos

# Imprime el recuento de outliers por cada columna
for col, proporcion in proporcion_outliers.items():
    print(f"Variable '{col}': {proporcion} outliers")
```

```python
# Pregunta1
# Cargar los datos
# Contar el n√∫mero de videos por categor√≠a
category_counts = dataset['category'].value_counts()

# Crear un gr√°fico de barras
plt.figure(figsize=(12, 8))
category_counts.plot(kind='bar', color='skyblue')
plt.title('N√∫mero de Videos en Tendencia por Categor√≠a')
plt.xlabel('Categor√≠a')
plt.ylabel('Cantidad de Videos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Crear una tabla con el conteo de videos por categor√≠a
category_counts_df = category_counts.reset_index()
category_counts_df.columns = ['Category', 'Number of Videos']

# Mostrar la tabla
print(category_counts_df.head(10))
```

```python
#Pregunta 2
category_likes = dataset.groupby('category')['likes'].sum().sort_values(ascending=False)
category_dislikes = dataset.groupby('category')['dislikes'].sum().sort_values(ascending=False)

# Creando gr√°ficos de barras para 'likes' y 'dislikes'
plt.figure(figsize=(12, 8))
category_likes.plot(kind='bar', color='green')
plt.title('Total de "Me gusta" por Categor√≠a')
plt.xlabel('Categor√≠a')
plt.ylabel('Total de "Me gusta"')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 8))
category_dislikes.plot(kind='bar', color='red')
plt.title('Total de "No me gusta" por Categor√≠a')
plt.xlabel('Categor√≠a')
plt.ylabel('Total de "No me gusta"')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Creando tablas para 'likes' y 'dislikes'
likes_df = category_likes.reset_index()
likes_df.columns = ['Category', 'Total Likes']

dislikes_df = category_dislikes.reset_index()
dislikes_df.columns = ['Category', 'Total Dislikes']

# Mostrar tablas
print(likes_df.head(10))
print(dislikes_df.head(10))
```

```python
#pregunta 3
# Calculando la proporci√≥n de "Me gusta" / "No me gusta" para cada categor√≠a
category_like_dislike_ratio = dataset.groupby('category').apply(lambda x: x['likes'].sum() / (x['dislikes'].sum() + 1))
category_like_dislike_ratio = category_like_dislike_ratio.sort_values(ascending=False)

# Creando gr√°fico de barras para la proporci√≥n de "Me gusta" / "No me gusta"
plt.figure(figsize=(12, 8))
category_like_dislike_ratio.plot(kind='bar', color='purple')
plt.title('Proporci√≥n de "Me gusta" / "No me gusta" por Categor√≠a')
plt.xlabel('Categor√≠a')
plt.ylabel('Ratio de "Me gusta" / "No me gusta"')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Creando tabla con la proporci√≥n de "Me gusta" / "No me gusta"
ratio_df = category_like_dislike_ratio.reset_index()
ratio_df.columns = ['Category', 'Like/Dislike Ratio']

# Mostrar tabla
print(ratio_df.head(10))
```

```python
#pregunta 4
# Calculando la proporci√≥n de "Vistas" / "Comentarios" para cada categor√≠a
category_view_comment_ratio = dataset.groupby('category').apply(lambda x: x['views'].sum() / (x['comment_count'].sum() + 1))
category_view_comment_ratio = category_view_comment_ratio.sort_values(ascending=False)

# Creando gr√°fico de barras para la proporci√≥n de "Vistas" / "Comentarios"
plt.figure(figsize=(12, 8))
category_view_comment_ratio.plot(kind='bar', color='orange')
plt.title('Proporci√≥n de "Vistas" / "Comentarios" por Categor√≠a')
plt.xlabel('Categor√≠a')
plt.ylabel('Ratio de "Vistas" / "Comentarios"')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Creando tabla con la proporci√≥n de "Vistas" / "Comentarios"
view_comment_ratio_df = category_view_comment_ratio.reset_index()
view_comment_ratio_df.columns = ['Category', 'View/Comment Ratio']

# Mostrar tabla
print(view_comment_ratio_df.head(10))
```

```python
#Pregunta 5
# Convertir la columna 'trending_date' a formato de fecha
dataset['trending_date'] = pd.to_datetime(dataset['trending_date'], format='%y.%d.%m')

# Agrupando los datos por fecha de tendencia y contando el n√∫mero de videos
trending_volume_over_time = dataset.groupby('trending_date').size() 

# Creando gr√°fico de l√≠neas para el volumen de videos en tendencia a lo largo del tiempo
plt.figure(figsize=(15, 8))
trending_volume_over_time.plot(kind='line', color='blue', marker='o')
plt.title('Volumen de Videos en Tendencia a lo Largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('N√∫mero de Videos en Tendencia')
plt.grid(True)
plt.tight_layout()
plt.show()

# Creando tabla para mostrar los datos
trending_volume_over_time_df = trending_volume_over_time.reset_index()
trending_volume_over_time_df.columns = ['Trending Date', 'Number of Trending Videos']

# Mostrar tabla
print(trending_volume_over_time_df.head(10))
```

```python
#pregunta 7
# Agrupando los datos por estado y calculando la suma de vistas, me gusta y no me gusta
state_views = dataset.groupby('state')['views'].sum().sort_values(ascending=False)
state_likes = dataset.groupby('state')['likes'].sum().sort_values(ascending=False)
state_dislikes = dataset.groupby('state')['dislikes'].sum().sort_values(ascending=False)

# Creando gr√°ficos de barras para vistas por estado
plt.figure(figsize=(12, 8))
state_views.head(10).plot(kind='bar', color='navy')
plt.title('Top 10 Estados con Mayor N√∫mero de Vistas')
plt.xlabel('Estado')
plt.ylabel('Total de Vistas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Creando tablas para vistas por estado
views_df = state_views.head(10).reset_index()
views_df.columns = ['State', 'Total Views']
print(views_df)

# Creando gr√°ficos de barras para me gusta por estado
plt.figure(figsize=(12, 8))
state_likes.head(10).plot(kind='bar', color='green')
plt.title('Top 10 Estados con Mayor N√∫mero de "Me gusta"')
plt.xlabel('Estado')
plt.ylabel('Total de "Me gusta"')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Creando tablas para me gusta por estado
likes_df = state_likes.head(10).reset_index()
likes_df.columns = ['State', 'Total Likes']
print(likes_df)

# Creando gr√°ficos de barras no me gusta por estado
plt.figure(figsize=(12, 8))
state_dislikes.head(10).plot(kind='bar', color='red')
plt.title('Top 10 Estados con Mayor N√∫mero de "No me gusta"')
plt.xlabel('Estado')
plt.ylabel('Total de "No me gusta"')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Creando tablas para no me gusta por estado
dislikes_df = state_dislikes.head(10).reset_index()
dislikes_df.columns = ['State', 'Total Dislikes']
print(dislikes_df)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

features = dataset[['views', 'dislikes', 'comment_count']]  # caracter√≠sticas
target = dataset['likes']  # Variable objetivo

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

# Evaluar el modelo
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print(f"Error cuadr√°tico medio (MSE): {mse}")
print(f"Coeficiente de determinaci√≥n (R^2): {r2}")
```
